# Autoencoders and multi-modality

## Introduction

In this project we will explore autoencoders and multi-modality:
1) An autoencoder is a deep neural network that can shrink down the dimensionality of input into a latent space representation (somewhat like a compression with some loss of data) with an encoder model and can reconstruct a close approximation of the original image with a decoder model.
2) A multi-modal neural network has the capability to swap parts of it in order to match our needs and perform different type of tasks.

We will try to use two approaches to create a multi-modal neural network, that can perform both image reconstruction and image classification with reasonable performance (given the amount of computational resources I have).

In this project we will be using PyTorch, which is one of the most popular deep learning libraries, and we will also utilize a Python package called "nn_utils" for some useful helper functions.

## Model building

We will build 3 different models that can work accomplish different tasks:
1) Encoder: This model can transform images into a latent-space representation that can be used by the other two models.
2) Decoder: The decoder can take a latent-space representation and construct an image that is an approximation of the original one, imputed to the encoder.
3) Classifier: This deep neural network can the latent-space representation of an image, created by the encoder and classify it by giving it a label.

The encoder and decoder use a ResNet-like architecture Conv2d(or ConvTranspose2d)-BarchNorm2d-LeakyReLU block combinations and residual connections, while the classifier uses a more traditional, fully connected architecture.

```python
class DownSampleBlock(nn.Module):
    def __init__(self, in_channels, out_channels, depth_factor=1, norm=True):
        assert depth_factor >= 1
        super().__init__()
        layers = [
            nn.Conv2d(in_channels, out_channels, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(out_channels, affine=True) if norm else nn.Identity(),
            nn.LeakyReLU(0.2, True)
        ]
        for _ in range(depth_factor - 1):
            layers += [
                nn.Conv2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1, bias=False),
                nn.BatchNorm2d(out_channels, affine=True) if norm else nn.Identity(),
                nn.LeakyReLU(0.2, True)
            ]
        self.layers = nn.Sequential(*layers)
        self.skip = nn.Conv2d(in_channels, out_channels, kernel_size=2, stride=2, padding=0, bias=False)

    def forward(self, x):
        s = x.clone()
        x = self.layers(x)
        x = x + self.skip(s)
        return x


class UpSampleBlock(nn.Module):
    def __init__(self, in_channels, out_channels, depth_factor=1, norm=True):
        assert depth_factor >= 1
        super().__init__()
        layers = [
            nn.ConvTranspose2d(in_channels, out_channels, kernel_size=4, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(out_channels, affine=True) if norm else nn.Identity(),
            nn.LeakyReLU(0.2, True)
        ]
        for _ in range(depth_factor - 1):
            layers += [
                nn.ConvTranspose2d(out_channels, out_channels, kernel_size=3, stride=1, padding=1, bias=False),
                nn.BatchNorm2d(out_channels, affine=True) if norm else nn.Identity(),
                nn.LeakyReLU(0.2, True)
            ]
        self.layers = nn.Sequential(*layers)
        self.skip = nn.ConvTranspose2d(in_channels, out_channels, kernel_size=2, stride=2, padding=0, bias=False)

    def forward(self, x):
        s = x.clone()
        x = self.layers(x)
        x = x + self.skip(s)
        return x


class Encoder(torch.nn.Module):
    def __init__(self, img_size, latent_size, depth_factor=1, width_factor=1):
        assert depth_factor >= 1 and width_factor >= 1
        super().__init__()
        num_channels = [16] + [2 ** (4 + i) * width_factor for i in range(int(np.log2(img_size)))]
        self.head = nn.Conv2d(3, num_channels[0], kernel_size=3, stride=1, padding=1, bias=False)
        self.blocks = nn.Sequential(
            *[DownSampleBlock(num_channels[i - 1], num_channels[i], depth_factor) for i in range(1, len(num_channels) - 1)] +
             [DownSampleBlock(num_channels[-2], num_channels[-1], depth_factor, norm=False)]
        )
        self.tail = nn.Sequential(
            nn.Linear(num_channels[-1], latent_size),
            nn.LeakyReLU(0.2, True)
        )

    def forward(self, x):
        x = self.head(x)
        x = self.blocks(x)
        x = self.tail(x[..., 0, 0])
        return x


class Decoder(torch.nn.Module):
    def __init__(self, img_size, latent_size, depth_factor=1, width_factor=1):
        super().__init__()
        self.latent_size = latent_size
        num_channels = [16] + [2 ** (4 + i) * width_factor for i in range(int(np.log2(img_size)))]
        num_channels = num_channels[::-1]
        self.head = nn.Sequential(
            nn.Linear(latent_size, num_channels[0]),
            nn.LeakyReLU(0.2, True)
        )
        self.blocks = nn.Sequential(
            *[UpSampleBlock(num_channels[i - 1], num_channels[i], depth_factor) for i in range(1, len(num_channels) - 1)] +
            [UpSampleBlock(num_channels[-2], num_channels[-1], depth_factor, norm=False)]
        )
        self.tail = nn.Sequential(
            nn.ConvTranspose2d(num_channels[-1], 3, kernel_size=3, stride=1, padding=1, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        x = self.head(x)
        x = self.blocks(x[..., None, None])
        x = self.tail(x)
        return x


class Classifier(nn.Sequential):
    def __init__(self, latent_size, num_classes):
        super().__init__(
            nn.Linear(latent_size, latent_size),
            nn.LeakyReLU(0.2, True),
            nn.Linear(latent_size, latent_size),
            nn.LeakyReLU(0.2, True),
            nn.Linear(latent_size, num_classes)
        )
```

## Approach 1: Autoencoder-first training

In this approach we will try to train the autoencoder (encoder + decoder) on its own to achieve a good performance and as a second step, we will freeze the encoder and attach an untrained classifier and train the classifier to use the output of the encoder for the classification task.

We will also perform some hyperparameter tuning with the autoencoder training so we can use it in the second approach as well.

From the results, we can see that the encoder-decoder pair perform well enough. They can compress the image into compact representation and reconstruct an image that looks like a slightly blured version of the original image, but the performance of the classifier does not meet expectations.

```python
def train_autoencoder(encoder, decoder, opt, train_loader, val_loader, scheduler=None, num_epochs=1, verbose=2):
    assert num_epochs >= 0
    assert next(encoder.parameters()).device == next(decoder.parameters()).device
    device = next(encoder.parameters()).device
    dtype = next(encoder.parameters()).dtype
    loss_fn = nn.MSELoss()
    num_batches = len(train_loader)

    for epoch_idx in range(num_epochs):
        if verbose >= 1:
            print(f'{epoch_idx + 1}/{num_epochs} epoch:')
        encoder.train()
        decoder.train()
        train_losses = []
        val_losses = []
        for batch_idx, (x, _) in enumerate(train_loader):
            x = x.to(device=device, dtype=dtype)
            opt.zero_grad()
            output = decoder(encoder(x))
            loss = loss_fn(output, x)
            loss.backward()
            opt.step()
            train_losses.append(loss.item())
            if verbose >= 2:
                print(f'\t{batch_idx + 1}/{num_batches} batch - training loss: {np.mean(train_losses):.6f}')
        with torch.no_grad():
            if scheduler is not None:
                scheduler.step()
            encoder.eval()
            decoder.eval()
            for x, _ in iter(val_loader):
                x = x.to(device=device, dtype=dtype)
                output = decoder(encoder(x))
                loss = loss_fn(output, x)
                val_losses.append(loss.item())
            if verbose >= 1:
                print(f'training loss: {np.mean(train_losses):.6f}, validation loss: {np.mean(val_losses):.6f}')
```

```python
encoder = Encoder(32, 256, depth_factor=1, width_factor=10).to(device=device, dtype=dtype)
decoder = Decoder(32, 256, depth_factor=1, width_factor=10).to(device=device, dtype=dtype)

opt = torch.optim.AdamW(itertools.chain(encoder.parameters(), decoder.parameters()), lr=0.0005, weight_decay=0.0001)
def lr_fn(epoch):
    if epoch < 10:  # [1, 10)
        return 1.
    if epoch < 15:  # [10, 15)
        return 1e-1
    return 1e-2  # [15, inf)
scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_fn)

train_autoencoder(encoder, decoder, opt, train_loader, val_loader, scheduler, num_epochs=20, verbose=1)
```

```python
classifier = Classifier(256, 10).to(device=device, dtype=dtype)
combined_model = nn.Sequential(encoder, classifier).to(device=device, dtype=dtype)
# Freezes encoder parameters to only train the classifier part
for p in combined_model[0].parameters():
    p.requires_grad = False

opt = torch.optim.AdamW(combined_model.parameters(), lr=0.0005, weight_decay=0.0001)

def lr_fn(epoch):
    if epoch < 25:  # [1, 25)
        return 1.
    if epoch < 40:  # [25, 40)
        return 1e-1
    return 1e-2  # [40, inf)
scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_fn)

train_classifier(combined_model, nn.CrossEntropyLoss(), opt,
                 train_loader, val_loader,
                 callbacks=[
                     #CLILoggerCallback(),  # This line is disabled because it produces a lot of text. Uncomment, if you want to see.
                     LRSchedulerCallback(scheduler)
                 ],
                 num_epochs=50)
```

## Approach 2: Multi-modal training

In this second approach we will attempt to train all three of the models simultaneously by trying to minimise the total sum of losses across all tasks. Here are some observations:
- The hyperparameters for the encoder and decoder from the previous approach are a good starting point but not perfect.
- Since the decoder uses MSE loss and the classifier uses cross-entropy loss, their scale is not the same. The expected starting value of the cross-entropy loss is log(10), which is much higher than the expected starting values of the MSE loss. If we add them, it is clear that in order to minimise the total loss, it is worth prioritising the classification loss more. This can lead to a bias between the tasks. To mitigate this, there is a weight for the losses of the different tasks.
- This approach requires more computational resources during training since all three models have to be kept in memory and computed with every forward and backward pass.

This approach presents a tradeoff. Although the classification accuracy is much better, the image reconstruction is worse compared to the previous approach.

```python
def train_multimodal(encoder, decoder, classifier, opt, train_loader, val_loader,
                     scheduler=None, loss_weights=(1., 1.), num_epochs=1, verbose=2):
    assert num_epochs >= 0
    assert next(encoder.parameters()).device == next(decoder.parameters()).device
    assert next(encoder.parameters()).device == next(classifier.parameters()).device
    device = next(encoder.parameters()).device
    dtype = next(encoder.parameters()).dtype
    # "r" stands for reconstruction
    # "c" stands for classification
    r_weight, c_weight = loss_weights
    r_loss_fn, c_loss_fn = nn.MSELoss(), nn.CrossEntropyLoss()
    num_batches = len(train_loader)

    for epoch_idx in range(num_epochs):
        if verbose >= 1:
            print(f'{epoch_idx + 1}/{num_epochs} epoch:')
        encoder.train()
        decoder.train()
        train_r_losses, train_c_losses = [], []
        val_r_losses, val_c_losses = [], []
        for batch_idx, (x, y) in enumerate(train_loader):
            x = x.to(device=device, dtype=dtype)
            y = y.to(device=device)
            opt.zero_grad()
            latent_vector = encoder(x)
            r_loss = r_weight * r_loss_fn(decoder(latent_vector), x)
            c_loss = c_weight * c_loss_fn(classifier(latent_vector), y)
            loss = r_loss + c_loss
            loss.backward()
            opt.step()
            train_r_losses.append(r_loss.item())
            train_c_losses.append(c_loss.item())
            if verbose >= 2:
                print(f'\t{batch_idx + 1}/{num_batches} batch - {np.mean(train_r_losses):.6f}, {np.mean(train_c_losses):.6f}')
        with torch.no_grad():
            if scheduler is not None:
                scheduler.step()
            encoder.eval()
            decoder.eval()
            for x, y in iter(val_loader):
                x = x.to(device=device, dtype=dtype)
                y = y.to(device=device)
                latent_vector = encoder(x)
                r_loss = r_weight * r_loss_fn(decoder(latent_vector), x)
                c_loss = c_weight * c_loss_fn(classifier(latent_vector), y)
                val_r_losses.append(r_loss.item())
                val_c_losses.append(c_loss.item())
            if verbose >= 1:
                print(
                    f'training losses: {np.mean(train_r_losses):.6f}, {np.mean(train_c_losses):.6f}\nvalidation losses: {np.mean(val_r_losses):.6f}, {np.mean(val_c_losses):.6f}'
                )
```

```python
encoder = Encoder(32, 256, depth_factor=2, width_factor=10).to(device=device, dtype=dtype)
decoder = Decoder(32, 256, depth_factor=2, width_factor=10).to(device=device, dtype=dtype)
classifier = Classifier(256, 10).to(device=device, dtype=dtype)

opt = torch.optim.AdamW(
    itertools.chain(encoder.parameters(), decoder.parameters(), classifier.parameters()),
    lr=0.0005, weight_decay=0.0001
)
def lr_fn(epoch):
    if epoch < 10:  # [1, 10)
        return 1.
    if epoch < 15:  # [10, 15)
        return 1e-1
    return 1e-2  # [15, inf)
scheduler = torch.optim.lr_scheduler.LambdaLR(opt, lr_fn)

train_multimodal(encoder, decoder, classifier, opt, train_loader, val_loader, scheduler, (18., 1.), num_epochs=20, verbose=1)
```

## Conclusion and discussion

Both approaches have their positives and negatives. The first one can reconstruct images fat better, while the second can better classify images. This tradeoff is probably due to the fact that the 256 dimensional latent-space representation is not enough to encode the images in a way that contains enough information to reconstruct good quality images and classify them as well. With a higher latent-space dimension size and a more powerful encode-decoder pair we might be able to achieve far better results.

In conclusion, the second approach may be better because the classification accuracy can reach 80% and higher and the image reconstruction is still somewhat working, although it is blurry.
